**Label Ecoding**

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DataSet ML/iris.csv')
print(df['species'].unique())
# print(df)
```

**categorical value will replace with the numerical value[int]**

```python

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['species']= label_encoder.fit_transform(df['species'])#Label in column 'species'
df['species'].unique()
```

**Binarization**

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import Binarizer


data = load_iris(as_frame=True)#Loading dataset
# df = pd.DataFrame(data.data, columns=data.feature_names)
df = data.frame
# df['species'] = data.target
df['sepal length (cm)']
```

**After Applying Binarization**

```python

binz = Binarizer(threshold=5)
df['sepal length (cm)'] = binz.fit_transform(df[['sepal length (cm)']])
df['sepal length (cm)']
```
